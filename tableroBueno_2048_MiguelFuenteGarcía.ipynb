```python
#Clase celda, objeto donde se encuentran todos los datos importantes sobre las casillas del tablero del 2048
class Celda:
    #lista de valores, que puede tomar la celda, siendo E cero, y las demas en orden alfabetico quitando la E
    listaValores=["E","A","B","C","D","F","G","H","I","J","K","L"]
    #Constructor de la clase, se dan por defecto los valores x y introducido por el constructor, la variable valor, 
    #es el valor que tiene la celda en cada momento, al crearla es E porque el tablero se crea vacio. El exponente es
    # el indice de la lista. 
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.valor="E"
        self.exponente=0
    # metodo de la celda que se realiza cuando 2 celdas con el mismo valor se juntan. Aumentamos en 1 el exponente y luego
    # escogemos el valor de la lista con el indice del exponente.
    def sumar(self):
        self.exponente=self.exponente+1
        self.valor=self.listaValores[self.exponente]
    # metodo auxiliar del sumar, que pone a 0 la celda que corresponda.
    def eliminar(self):
        self.exponente=0
        self.valor=self.listaValores[self.exponente]
    # metodo que cambia los valores entre 2 celdas contiguas.
    def cambiar(self, celda):
        celdaAdyacente=celda
        aux=celdaAdyacente.valor
        celdaAdyacente.valor=self.valor
        self.valor=aux
        exp=celdaAdyacente.exponente
        celdaAdyacente.exponete=self.exponente
        self.exponente=exp
        aux=0
        exp=0
    def __eq__(self, other) -> bool:
        return self.x==other.x & self.y==other.y
```

```python
import random
import copy
# Clase tablero, contiene todos los atributos de mover y añadir.
class Tablero:
    # constructor, inicia las dimensiones del tablero que son 4x4, se crea una matriz de celdas(objeto) y una matriz
    #que contiene los valores de las celdas.
    def __init__(self):
        self.alto=4
        self.ancho=4
        self.tablero=[]
        self.tableroVisual=[]
        for i in range(self.alto):
            self.tablero.append([])
            self.tableroVisual.append([])
            for j in range(self.ancho):
                self.tablero[i].append(Celda(i,j))
                self.tableroVisual[i].append(self.tablero[i][j].valor)   
            
    # metodo que devuelve las coordenadas de las celdas cuyo valor es "E" o 0    
    def listaCeldasVacias(self, tableroAnalizar):
    # lista de celdas vacias
        listaVacia=[]
        for i in range(self.alto):
            for j in range(self.ancho):
                if(tableroAnalizar.tablero[i][j].valor=="E"):
                    listaVacia.append([i,j])
        return listaVacia
    #metodo que comprueba si una fila esta toda vacia
    def fila_esta_libre(self, fila):
        for j in range (self.alto):
            if(self.tablero[fila][j].valor=="E"):
                continue
            else:
                return False
        return True
    #metodo que comprueba si una columna esta vacia
    def columna_esta_libre(self, columna):
        for j in range (self.ancho):
            if(self.tablero[j][columna].valor=="E"):
                continue
            else:
                return False
        return True
    
    def tableroVisible(self):
        for i in range(self.alto):
            for j in range(self.ancho):
                self.tableroVisual[i][j]=self.tablero[i][j].valor
        return self.tableroVisual
    #metodo que permite mover a la izquierda el tablero que se le da como atributo
    def izquierda(self, tableroInterno):
    
        for i in range(self.alto):
            #comprueba si la fila esta libre, porque en ese caso, no va a mover nada.
            if(self.fila_esta_libre(i)):
                continue
            else:
                #contador que va a contar los 0 que ha encontrado en la fila
                self.contador=0
                for j in range(self.ancho):
                    # si el valor es 0, movemos ese 0 a la derecha del todo de la fila
                    if(tableroInterno[i][j-self.contador].valor=="E"):
                        self.x=j-self.contador
                        #bucle que permite cambiar los valores de izquierda a derecha de 2 en 2, es decir
                        # en el caso de que sea 0 2 4 2 va ha hacer: 2 0 4 2-> 2 4 0 2 -> 2 4 2 0
                        while(self.x+1<4):
                            tableroInterno[i][self.x].cambiar(tableroInterno[i][self.x+1])
                            self.x=self.x+1
                            
                        self.contador=self.contador+1
                    else:
                        #En caso de que haya un numero distinto de 0
                        
                        #para que no se referencie a una celda <0
                        if(j-self.contador-1>=0):
                            
                            # si el valor de la celda actual es igual al que la celda de su izquierda se suman

                            if(tableroInterno[i][j-self.contador].valor==tableroInterno[i][j-self.contador-1].valor):
                                #suma en uno la celda de la izquierda
                                tableroInterno[i][j-self.contador-1].sumar() 
                                #elimina el numero de la derecha (pone a 0)
                                tableroInterno[i][j-self.contador].eliminar()
                                
                                self.x=j-self.contador
                                #bucle que pone a la derecha el 0 que acaba de crear al sumar los 2 numeros
                                while(self.x<3):
                                    tableroInterno[i][self.x].cambiar(tableroInterno[i][self.x+1])#elimina la celda
                                    self.x=self.x+1
                                #Un caso especial es que ocurra esto: 2 2 0 4 al sumar los 2 movera los 0 a la derecha
                                #4 0 4 0 quedando asi, luego realiza el movimiento 4 4 0 0 y en el siguiente movimiento 
                                #sumara 4 4 y no deberia.
                                #Por tanto, En caso de que haya sumado, una vez movido el 0 creado, si es un 0 y esta dentro de los limites
                                # de la matriz, cambia los valores de esas celdas. Y termina el bucle
                                if((tableroInterno[i][j-self.contador].valor=="E")&(j-self.contador+1<4)):
                                    tableroInterno[i][j-self.contador].cambiar(tableroInterno[i][j-self.contador+1])
                                    break
                            else:
                                continue
    #metodo que comprueba si se puede mover a la izquierda, comparando 2 tableros,
    # en el primero se realiza la accion de mover y el segundo no se mueve
    def puedeIzquierda(self):
        tableroaux=copy.deepcopy(self.tablero)
        tableroaux2=copy.deepcopy(tableroaux)
        self.izquierda(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroaux2[i][j].valor):
                    continue
                else:
                    return True        
        return False
    #metodo que permite mover a la derecha el tablero que se le da como atributo
    def derecha(self, tableroInterno):

        for i in range(self.alto):
            #comprueba si la fila esta libre, porque en ese caso, no va a mover nada.
            if(self.fila_esta_libre(i)):
                continue
            else:
                #contador que va a contar los 0 que ha encontrado en la fila
                self.contador=0
               
                for j in range(self.ancho):
                     # si el valor es 0, movemos ese 0 a la izquierda del todo de la fila
                    if(tableroInterno[i][self.ancho-j+self.contador-1].valor=="E"):
                        self.x=self.ancho-j+self.contador-1
                        #bucle que permite cambiar los valores de derecha a izquieda de 2 en 2, es decir
                        # en el caso de que sea 2 0 8 0 va ha hacer: 2 0 0 8-> 2 0 0 8 -> 0 2 0 8
                        while(self.x-1>=0):
                            tableroInterno[i][self.x].cambiar(tableroInterno[i][self.x-1])#elimina la celda
                            self.x=self.x-1
                            
                        self.contador=self.contador+1 
                    else:
                        #para que no se referencie a una celda >=4
                        if(self.ancho-j+self.contador<4):
                            #si la celda anterior esta dentro de los limites

                            if(tableroInterno[i][(self.ancho-j+self.contador)].valor==tableroInterno[i][self.ancho-j+self.contador-1].valor):
                                #suma en uno la celda de la derecha
                                tableroInterno[i][self.ancho-j+self.contador].sumar() 
                                #elimina el numero de la izquierda (pone a 0)
                                tableroInterno[i][self.ancho-j+self.contador-1].eliminar()
                                self.x=self.ancho-j+self.contador-1
                                #bucle que pone a la izquierda el 0 que acaba de crear al sumar los 2 numeros
                                while(self.x>0):
                                    tableroInterno[i][self.x].cambiar(tableroInterno[i][self.x-1])
                                    self.x=self.x-1
                                #Para evitar el caso especial comentado en el movimiento Izquierda
                                if((tableroInterno[i][self.ancho-j+self.contador-1].valor=="E")&(self.ancho-j+self.contador-1>0)):
                                    tableroInterno[i][self.ancho-j+self.contador-1].cambiar(tableroInterno[i][self.ancho-j+self.contador-2])
                                    break
                            else:
                                continue
                                
                                
    #metodo que comprueba si se puede mover a la derecha, comparando 2 tableros,
    # en el primero se realiza la accion de mover y el segundo no se mueve
    def puedeDerecha(self):
        tableroaux=copy.deepcopy(self.tablero)
        tableroaux2=copy.deepcopy(tableroaux)
        self.derecha(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroaux2[i][j].valor):
                    continue
                else:
                    return True        
        return False
    #metodo que permite mover hacia arriba el tablero que se le da como atributo
    def arriba(self, tableroInterno):

        for i in range(self.ancho):
            #comprueba si la columna esta libre, porque en ese caso, no va a mover nada.
            if(self.columna_esta_libre(i)):
                continue

            else:
                #contador que va a contar los 0 que ha encontrado en la fila
                self.contador=0  
                
                for j in range(self.alto):
                     # si el valor es 0, movemos ese 0 abajo del todo de la columna
                    if(tableroInterno[j-self.contador][i].valor=="E"):
                            
                            self.x=j-self.contador
                            while(self.x+1<4):
                                tableroInterno[self.x][i].cambiar(tableroInterno[self.x+1][i])#elimina la celda
                                self.x=self.x+1
                                
                            self.contador=self.contador+1 
                    else: 
                        #para que no se referencie a una celda <0
                        if(j-self.contador-1>=0):
                            #si el valor de la celda es igual al de la siguiente
                            if(tableroInterno[j-self.contador][i].valor==tableroInterno[j-self.contador-1][i].valor):
                                #sumamos el numero de arriba y borramos el de abajo
                                    tableroInterno[j-self.contador-1][i].sumar()
                                    tableroInterno[j-self.contador][i].eliminar()
                                    self.x=j-self.contador
                                    #bucle que pone abajo el 0 que acaba de crear al sumar los 2 numeros
                                    while(self.x<3):
                                        tableroInterno[self.x][i].cambiar(tableroInterno[self.x+1][i])
                                        self.x=self.x+1
                                    #Para evitar el caso especial comentado en el movimiento Izquierda
                                    if((tableroInterno[j-self.contador][i].valor=="E")&(j-self.contador+1<4)):
                                        tableroInterno[j-self.contador][i].cambiar(tableroInterno[j-self.contador+1][i])
                                        break
                            else:
                                continue
                                
                                
    #metodo que comprueba si se puede mover hacia arriba, comparando 2 tableros,
    # en el primero se realiza la accion de mover y el segundo no se mueve
    def puedeArriba(self):
        tableroaux=copy.deepcopy(self.tablero)
        tableroaux2=copy.deepcopy(tableroaux)
        self.arriba(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroaux2[i][j].valor):
                    continue
                else:
                    return True        
        return False
    
    #metodo que permite mover hacia abajo el tablero que se le da como atributo
    def abajo(self, tableroInterno):
    
        for i in range(self.ancho):
             #comprueba si la columna esta libre, porque en ese caso, no va a mover nada.
            if(self.columna_esta_libre(i)):
                continue
            else:
                #contador que va a contar los 0 que ha encontrado en la fila
                self.contador=0
                for j in range(self.alto):
                     # si el valor es 0, movemos ese 0 arriba del todo de la columna
                    if(tableroInterno[self.alto-j+self.contador-1][i].valor=="E"):
                        
                        self.x=self.alto-j+self.contador-1
                        while(self.x-1>=0):
                            tableroInterno[self.x][i].cambiar(tableroInterno[self.x-1][i])#elimina la celda
                            self.x=self.x-1
                        self.contador=self.contador+1 
                    else:
                        #para que no se referencie a una celda >=4
                        if(self.alto-j+self.contador<4):


                            if(tableroInterno[(self.alto-j+self.contador)][i].valor==tableroInterno[self.alto-j+self.contador-1][i].valor):
                                #sumamos el numero de abajo y borramos el de rriba
                                tableroInterno[self.alto-j+self.contador][i].sumar() #(En este caso) Multiplica a uno de las 2 celdas por 2 y elimina la otra
                                tableroInterno[self.alto-j+self.contador-1][i].eliminar()# aqui eliminamos el numero
                                self.x=self.alto-j+self.contador-1
                                #bucle que pone arriba el 0 que acaba de crear al sumar los 2 numeros
                                while(self.x>0):
                                    tableroInterno[self.x][i].cambiar(tableroInterno[self.x-1][i])#elimina la celda
                                    self.x=self.x-1
                                 #Para evitar el caso especial comentado en el movimiento Izquierda
                                if((tableroInterno[self.alto-j+self.contador-1][i].valor=="E")&(j-self.contador-1>0)):
                                        tableroInterno[self.alto-j+self.contador-1][i].cambiar(tableroInterno[self.alto-j+self.contador-2][i])
                                        break
                            else:
                                continue
                                
                                
    #metodo que comprueba si se puede mover hacia abajo, comparando 2 tableros,
    # en el primero se realiza la accion de mover y el segundo no se mueve
    def puedeAbajo(self):
        tableroaux=copy.deepcopy(self.tablero)
        tableroaux2=copy.deepcopy(tableroaux)
        self.abajo(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroaux2[i][j].valor):
                    continue
                else:
                    return True        
        return False
    
    #metodo que comprueba si el objeto es igual a otro objeto
    def __eq__(self, o: object) -> bool:
        for i in self.ancho:
            for j in self.alto:
                if(self.tablero[i][j].valor==object.tablero[i][j].valor):
                    continue
                else:
                    return False
        return True
    
    #metodo que permite añadir al tablero numeros, despues de cada movimiento
    # siendo 2 un 90% de las veces y 4 un 10%
    def añadir(self,lista):
        numeroMax=len(lista)-1
        numeroListaRd=random.randint(0,numeroMax)
        if(random.randint(0,10)==0):
            self.tablero[lista[numeroListaRd][0]][lista[numeroListaRd][1]].valor="B"
            self.tablero[lista[numeroListaRd][0]][lista[numeroListaRd][1]].exponente=2
        else:
            self.tablero[lista[numeroListaRd][0]][lista[numeroListaRd][1]].valor="A"
            self.tablero[lista[numeroListaRd][0]][lista[numeroListaRd][1]].exponente=1
    
    #metodo que devuelve el numero del movimiento que se tiene que hacer para llegar a ese hijo.
    def getMoveTo(self, tableroOptimo)-> int:
        #creamos tableros auxiliares para que pueda realizar los movimientos y ver si es el hijo
        #hacemos eso para los 4 movimientos
        tableroaux=copy.deepcopy(self.tablero)
        self.arriba(tableroaux)
        valordistinto=False
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroOptimo[i][j].valor):
                    continue
                else:
                    valordistinto=True
        if(valordistinto==False):
            return 2
        valordistinto=False
        tableroaux=copy.deepcopy(self.tablero)
        self.abajo(tableroaux)
        
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroOptimo[i][j].valor):
                    continue
                else:
                    valordistinto=True
        if(valordistinto==False):
            return 3
        valordistinto=False
        tableroaux=copy.deepcopy(self.tablero)
        self.izquierda(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroOptimo[i][j].valor):
                    continue
                else:
                    valordistinto=True
        if(valordistinto==False):
            return 0   
        valordistinto=False
        tableroaux=copy.deepcopy(self.tablero)
        self.derecha(tableroaux)
        for i in range(4):
            for j in range(4):
                if(tableroaux[i][j].valor==tableroOptimo[i][j].valor):
                    continue
                else:
                    valordistinto=True
        if(valordistinto==False):
            return 1

        
```

```python
from IPython.display import display
from IPython.display import clear_output
from ipywidgets import HTML, Dropdown ,Button, HBox, VBox, Box
import numpy as np
import math


#Clase arbitro, contiene el main y otros metodos para controlar el tablero.
class Arbitro:
    #constructor, instancia al tablero, e introduce variables que se usan en el tablero.
    def __init__(self):
        self.lose=False
        self.miTablero=Tablero()
        self.puntos=0
        self.celdasOcupadas=0
        self.botonPulsado=False
        self.primero=True
        self.movRecomendado=0
    # metodo equal, que compara cuando 2 arbitros son iguales
    def __eq__(self,other) ->bool:
        for i in self.miTablero.alto:
            for j in self.miTablero.ancho:
                if(self.miTablero.tablero[i][j]==(other.miTablero.tablero[i][j])):
                    continue
                else:
                    return False
        return True
    
    #metodo utilidad, que devuelve la funcion heuristica del programa
    def utility(self,tableroHeu):
        matriz=[]
        for i in range(tableroHeu.alto):
            matriz.append([])
            for j in range(tableroHeu.ancho):
                matriz[i].append(2**tableroHeu.tablero[i][j].listaValores.index(tableroHeu.tablero[i][j].valor))
        matrizNumpy=np.array(matriz)
        
        #Esta matriz, permite dar prioridad a los movimientos hacia arriba antes que a la izquierda.
        #Esta funcion es mejor que la de monotonia porque no hay valores iguales
        matrizSnake=[ [2**15  ,2**14,2**13,2**12],  
                     [2**8 , 2**9,2**10,2**11], 
                     [2**7, 2**6,2**5,2**4],
                     [2**0,2**1,2**2,2**3]]
        #en caso de que haya una celda con 2048, que el valor que devuelva la heuristica se multiplique por 8, para 
        # que haya mas probabilidades de llegar
        if(np.max(matrizNumpy)==2048):
            return np.sum(np.multiply(matrizNumpy,matrizSnake))*8
        
        # en caso de que el mayor de los numeros se encuentre en la celda de arriba a la izquierda
        if(np.max(matrizNumpy)==2**tableroHeu.tablero[0][0].listaValores.index(tableroHeu.tablero[0][0].valor)):
            return np.sum(np.multiply(matrizNumpy,matrizSnake))
        else:
            #en caso contrario el resultado entre 8 para que haya menos probabilidad de que se obtenga este resultado
            return np.sum(np.multiply(matrizNumpy,matrizSnake))/8
        
    # metodo que devuelve los movimientos que puede hacer el tablero adjuntado
    def movimientos(self,tablero):
        listaMovimientos=[]
        listaMovimientos.clear
        if(tablero.puedeIzquierda()==True):
            listaMovimientos.append(0)
        if(tablero.puedeDerecha()==True):
            listaMovimientos.append(1)
        if(tablero.puedeArriba()==True):
            listaMovimientos.append(2)
        if(tablero.puedeAbajo()==True):
            listaMovimientos.append(3)
        return listaMovimientos
    
    # metodo que comprueba si se ha ganado la partida, es decir si hay 2048 en una de las celdas
    def hasGanado(self):
        for i in range(self.miTablero.alto):
            for j in range(self.miTablero.ancho):
                if(self.miTablero.tablero[i][j].valor=='L'):
                    return True
                
        return False
    
    # metodo que permite mover el tablero, cuando se pulsa un boton
    def on_button_clicked(self,b):
        desc=b.description
        # se obtienen los movimientos que puede hacer el tablero
        self.posiblesMovimientos=self.movimientos(self.miTablero)
        self.primero=True
        self.botonPulsado=True
        
        #si el boton pulsado, tiene la descripcion ^ mueve para arriba el tablero
        if desc=="^":
            self.botonNumero=2
            #se comprueba si el boton pulsado esta en la lista de movimientos disponibles, si esta lo mueve
            #si no, no realiza ningun movimiento
            
            if(self.botonNumero in self.posiblesMovimientos):
                self.miTablero.arriba(self.miTablero.tablero)
                self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                self.posiblesMovimientos=self.movimientos(self.miTablero)
                if(len(self.posiblesMovimientos)==0):
                    print("Has perdido")
                    return 0
            else:
                print("No se puede realizar ese movimiento")
        #si el boton pulsado, tiene la descripcion v mueve para abajo el tablero
        elif desc=="v":
            self.botonNumero=3
            
            #se comprueba si el boton pulsado esta en la lista de movimientos disponibles, si esta lo mueve
            #si no, no realiza ningun movimiento
            
            if(self.botonNumero in self.posiblesMovimientos):
                self.miTablero.abajo(self.miTablero.tablero)
                self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                self.posiblesMovimientos=self.movimientos(self.miTablero)

                if(len(self.posiblesMovimientos)==0):
                    print("Has perdido")
                    return 0
            else:
                 print("No se puede realizar ese movimiento")
        #si el boton pulsado, tiene la descripcion > mueve para la derecha el tablero
        elif desc==">":
            self.botonNumero=1
            
            #se comprueba si el boton pulsado esta en la lista de movimientos disponibles, si esta lo mueve
            #si no, no realiza ningun movimiento
            
            if(self.botonNumero in self.posiblesMovimientos):
                self.miTablero.derecha(self.miTablero.tablero)
                self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                self.posiblesMovimientos=self.movimientos(self.miTablero)
                if(len(self.posiblesMovimientos)==0):
                    print("Has perdido")
                    return 0
            else:
                print("No se puede realizar ese movimiento")
        #si el boton pulsado, tiene la descripcion > mueve para la izquierda el tablero
        elif desc=="<":
            self.botonNumero=0
            
            #se comprueba si el boton pulsado esta en la lista de movimientos disponibles, si esta lo mueve
            #si no, no realiza ningun movimiento
            
            if(self.botonNumero in self.posiblesMovimientos):
                self.miTablero.izquierda(self.miTablero.tablero)
                self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                

                self.posiblesMovimientos=self.movimientos(self.miTablero)
                if(len(self.posiblesMovimientos)==0):
                    print("Has perdido")
                    return 0

            else:
                 print("No se puede realizar ese movimiento")
        
        #limpia la pantalla
        clear_output()
        # y volvemos a imprimir el visor donde se encuentran los botones y el tablero
        up = Button(description="^")
        down = Button(description="v")
        right = Button(description=">")
        left = Button(description="<")
        empty = Button(description=" ")
        automatico = Button(description="Automatico")
        empty.margin=2
        self.visor=HTML()
        self.visor.value=self.get_html()
        control=VBox([HBox([empty,up,automatico]),HBox([left,down,right])])
        ui=VBox(children=[self.visor, control])
        display(ui)
        #Aqui termina de imprimir el tablero, y comprueba si se ha ganado la partida
        if(self.hasGanado()==True):
            print("Has Ganado")
            return 0
        #Llama a decision, que devuelve el movimiento que se recomienda hacer
        self.movRecomendado=self.decision()
        if self.movRecomendado==-1:
            print("No se puede hacer nada has perdido")
        if self.movRecomendado==0:
            print("Recomiendo izquierda")
        if self.movRecomendado==1:
            print("Recomiendo derecha")
        if self.movRecomendado==2:
            print("Recomiendo arriba")
        if self.movRecomendado==3:
            print("Recomiendo abajo")
        #Y se definen las acciones que hacen los botones
        up.on_click(self.on_button_clicked)
        down.on_click(self.on_button_clicked)
        right.on_click(self.on_button_clicked)
        left.on_click(self.on_button_clicked)
        self.botonPulsado=False
        automatico.on_click(self.automatico) 
        return 0
    
    #Metodo que se utiliza como main del arbitro, es donde se inicia  el 2048
    
    def main(self):
        #Añadimos 2 numeros, como indica el juego.
        self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
        self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
        self.posiblesMovimientos=self.movimientos(self.miTablero)
        self.primero=True
        # Declaramos los valores de los botones 
        up = Button(description="^")
        down = Button(description="v")
        right = Button(description=">")
        left = Button(description="<")
        empty = Button(description=" ")
        automatico = Button(description="Automatico")
        empty.margin=2
        #Ejecutamos el HTML, para que se puedan ver las imagenes del tablero
        self.visor=HTML()
        self.visor.value=self.get_html()
        #Imprimimos el tablero por pantalla
        if(self.primero==True): 
            control=VBox([HBox([empty,up,automatico]),HBox([left,down,right])])
            ui=VBox(children=[self.visor, control])
            display(ui)
            self.primero=False
        #Y recomendamos el movimiento a ejecutar antes de que pueda pulsar los botones
        self.movRecomendado=self.decision()
        if self.movRecomendado==-1:
            print("No se puede hacer nada has perdido")
        if self.movRecomendado==0:
            print("Recomiendo izquierda")
        if self.movRecomendado==1:
            print("Recomiendo derecha")
        if self.movRecomendado==2:
            print("Recomiendo arriba")
        if self.movRecomendado==3:
            print("Recomiendo abajo")
        if(len(self.posiblesMovimientos)==0):
            print("Has perdido")
        else:
            self.botonPulsado=False
            automatico.on_click(self.automatico)    
            up.on_click(self.on_button_clicked)
            down.on_click(self.on_button_clicked)
            right.on_click(self.on_button_clicked)
            left.on_click(self.on_button_clicked)
            
           
    #metodo que permite jugar al 2048 de forma automatica, para no tener que estar pulsando botones
    def automatico(self,b):
        while(True):
            desc=b.description
            #miramos los movimientos que puede realizar el tablero
            self.posiblesMovimientos=self.movimientos(self.miTablero)
            if(len(self.posiblesMovimientos)==0):
                    print("Has perdido")
                    return 0
            #que no vuelva a recomendar el movimiento en la primera ejecucion
            if(self.primero==False):
                self.movRecomendado=self.decision()
            else:
                self.primero=False
            
            #dependiendo del movimiento que recomiende, hacemos ese movimiento
            if self.movRecomendado==2:
                self.miTablero.arriba(self.miTablero.tablero)
                self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                #print(self.utility(self.miTablero))
            if self.movRecomendado==3:
                    self.miTablero.abajo(self.miTablero.tablero)
                    self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                    #print(self.utility(self.miTablero))
                    self.posiblesMovimientos=self.movimientos(self.miTablero)
            if self.movRecomendado==1:
                    self.miTablero.derecha(self.miTablero.tablero)
                    self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                    #print(self.utility(self.miTablero))
                    self.posiblesMovimientos=self.movimientos(self.miTablero)
            if self.movRecomendado==0:
                    self.miTablero.izquierda(self.miTablero.tablero)
                    self.miTablero.añadir(self.miTablero.listaCeldasVacias(self.miTablero))
                    #print(self.utility(self.miTablero))
                    self.posiblesMovimientos=self.movimientos(self.miTablero)
            #limpiamos la pantalla
            clear_output()
            #declaramos el tablero y los botones, para imprimirlos por pantalla.
            up = Button(description="^")
            down = Button(description="v")
            right = Button(description=">")
            left = Button(description="<")
            empty = Button(description=" ")
            automatico = Button(description="Automatico")
            empty.margin=2
            self.visor=HTML()
            self.visor.value=self.get_html()
            control=VBox([HBox([empty,up,automatico]),HBox([left,down,right])])
            ui=VBox(children=[self.visor, control])
            display(ui)
            #no declaramos los botones como acciones, porque el solo va a terminar la partida, para bien o para mal.
            if(self.hasGanado()==True):
                print("Has Ganado")
                return 0
            
        return 0
           
    #metodo que permite pasar de un tablero escrito a imagenes que pasa a su vez a HTML    
    def get_html(self):
        """ Muestra una representación gráfica del juego.

        Devuelve un "string" que contiene HTML

        """ 
        
        
        height = (self.miTablero.alto)
        width = (self.miTablero.ancho)
        
        html_string = "<style> img.game {width: 37px !important; height: 22px !important;}</style><table>"
        


        new_row = "<tr>"
        end_row = "</tr>"
        
        for i in range(height):
            html_string+=new_row
            for j in range(width):
                
                content = self.miTablero.tablero[i][j].valor
                drawing = elementoImagen[content]
                html = '<td><img class="game" src=%s alt=""></img></td>' % drawing     
                
                
                html_string+=html
            html_string+=end_row
                
        html_string += "</table>"
            
                    
        return html_string
    
    # metodo que llama al minimax, para que recomiende el mejor movimiento
    def decision(self):
        hijo=self.maximo(self.miTablero, 5,-math.inf,math.inf)
        return self.miTablero.getMoveTo(hijo.tablero)
        
        
    # metodo que obtiene el maximo de los 4 tableros hijos( a su vez llama a minimo, donde se calcula el minimo de los tableros)    
    def maximo(self,tablero, profundidad, alpha, beta):
        movimientosPosibles=self.movimientos(tablero)
        if (len(movimientosPosibles)==0):
                return self.utility(tablero)
        profundidad=profundidad-1
        score=-math.inf
        score_max=-math.inf
        for i in movimientosPosibles:
            if i==0:
                tableroPorBucle=copy.deepcopy(tablero)
                tableroPorBucle.izquierda(tableroPorBucle.tablero)
                if(profundidad==0):
                    score=self.utility(tableroPorBucle)
                else:
                    score=self.minimo(tableroPorBucle, profundidad,alpha,beta)
                if(score>score_max):
                    score_max=score
                    mejorTablero=tableroPorBucle
                if(score_max>=beta):
                    break
                if(score_max>alpha):
                    alpha=score_max
            if i==1:
                tableroPorBucle=copy.deepcopy(tablero)    
                tableroPorBucle.derecha(tableroPorBucle.tablero)
                if(profundidad==0):
                    score=self.utility(tableroPorBucle)
                else:
                    score=self.minimo(tableroPorBucle, profundidad,alpha,beta)
                if(score>score_max):
                    score_max=score
                    mejorTablero=tableroPorBucle
                if(score_max>=beta):
                    break
                if(score_max>alpha):
                    alpha=score_max
            if i==2:
                tableroPorBucle=copy.deepcopy(tablero)
                tableroPorBucle.arriba(tableroPorBucle.tablero)
                if(profundidad==0):
                    score=self.utility(tableroPorBucle)
                else:
                    score=self.minimo(tableroPorBucle, profundidad,alpha,beta)
                if(score>score_max):
                    score_max=score
                    mejorTablero=tableroPorBucle
                if(score_max>=beta):
                    break
                if(score_max>alpha):
                    alpha=score_max
            if i==3:
                tableroPorBucle=copy.deepcopy(tablero)
                tableroPorBucle.abajo(tableroPorBucle.tablero)
                if(profundidad==0):
                    score=self.utility(tableroPorBucle)
                else:
                    score=self.minimo(tableroPorBucle, profundidad,alpha,beta)
                if(score>score_max):
                    score_max=score
                    mejorTablero=tableroPorBucle
                if(score_max>=beta):
                    break
                if(score_max>alpha):
                    alpha=score_max
                    
        if(profundidad==4):
            #en caso de que la profundidad sea 4 en vez de devolver la heuristica del tablero, devuelve el tablero que tiene esa heuristica
            return(mejorTablero)
        else:
            return(score_max)
        
        
     #metodo que devuelve el minimo valor heuristico de los hijos del tablero que se le asigna, llama a maximo para complementar el minimax   
    def minimo(self,tablero, profundidad,alpha, beta):
        movimientosPosibles=self.movimientos(tablero)
        if (len(movimientosPosibles)==0):
                return self.utility(tablero)
                
        score_min=math.inf
        profundidad=profundidad-1
        if(profundidad!=0):
            listaceldasvacias=tablero.listaCeldasVacias(tablero)
            for i in listaceldasvacias:

                tablero.tablero[i[0]][i[1]].valor="A"
                tablero.tablero[i[0]][i[1]].exponente=1
                score=self.maximo(tablero,profundidad,alpha,beta)
                tablero.tablero[i[0]][i[1]].valor="E"
                tablero.tablero[i[0]][i[1]].exponente=0
                if(score<score_min):
                    score_min=score
                if(score_min<=alpha):
                    break
                if(score_min<beta):
                    beta=score_min
                #un caso que comentar en especial es este, porque teniendo en la misma celda el valor 2 y el 4,
                #siempre el valor 2 va a ser el minimo; por eso no se implementa esta opcion.
                #Se deja como comentario, por su posible uso en una mejora
                """tablero.tablero[i[0]][i[1]].valor="B"
                tablero.tablero[i[0]][i[1]].exponente=2
                score=self.maximo(tablero,profundidad,alpha,beta)
                tablero.tablero[i[0]][i[1]].valor="E"
                tablero.tablero[i[0]][i[1]].exponente=0
                if(score<score_min):
                    score_min=score
                if(score_min<=alpha):
                    break
                if(score_min<beta):
                    beta=score_min"""


            return score_min
                   
            
                
                
                
                    
                
    
    
```
